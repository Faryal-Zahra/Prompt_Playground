## Step 1: Load TinyLlama

```python
from transformers import pipeline

generator = pipeline(
    "text-generation",
    model="TinyLlama/TinyLlama-1.1B-Chat-v1.0"
)

response = generator(
    "What is Artificial Intelligence?",
    max_new_tokens=100
)

print(response[0]["generated_text"])
```

## Step 2: Create Reusable Function

```python
from transformers import pipeline

generator = pipeline(
    "text-generation",
    model="TinyLlama/TinyLlama-1.1B-Chat-v1.0"
)

def ask_llm(prompt):

    result = generator(
        prompt,
        max_new_tokens=150,
        temperature=0.7,
        do_sample=True
    )

    return result[0]["generated_text"]
```

## Test

```python
print(
    ask_llm("Explain Machine Learning")
)
```

## Step 4: Prompt Engineering Playground

```python
prompts = {

    "Basic":
    "Explain Machine Learning",

    "Teacher":
    """
    You are an expert teacher.

    Explain Machine Learning.
    """,

    "Child":
    """
    Explain Machine Learning to a 10 year old child.
    """,

    "Software Engineer":
    """
    Explain Machine Learning for software engineers.
    """,

    "Interview":
    """
    Explain Machine Learning for a technical interview.
    """
}
```

```python
for style, prompt in prompts.items():

    print("=" * 60)
    print(style)
    print("=" * 60)

    response = ask_llm(prompt)

    print(response)
    print("\n")
```
