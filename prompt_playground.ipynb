```python
from transformers import pipeline

generator = pipeline(
    "text-generation",
    model="gpt2"
)

def ask_llm(prompt):

    response = generator(
        prompt,
        max_new_tokens=100,
        temperature=0.7,
        do_sample=True
    )

    return response[0]["generated_text"]
```

```python
print(
    ask_llm(
        "Explain Machine Learning"
    )
)
```

```python
prompts = {

    "Basic":
    "Explain Machine Learning",

    "Teacher":
    """
    You are an expert teacher.

    Explain Machine Learning.
    """,

    "Child":
    """
    Explain Machine Learning to a 10 year old child.
    """,

    "Software Engineer":
    """
    Explain Machine Learning for software engineers.
    """,

    "Interview":
    """
    Explain Machine Learning for a technical interview.
    """
}
```

```python
for style, prompt in prompts.items():

    print("=" * 60)
    print(f"STYLE: {style}")
    print("=" * 60)

    result = ask_llm(prompt)

    print(result)
    print("\n\n")
```

```python
import pandas as pd

results = []

for style, prompt in prompts.items():

    answer = ask_llm(prompt)

    results.append({
        "Prompt Style": style,
        "Response": answer
    })

df = pd.DataFrame(results)

df
```

```python
df.to_csv(
    "prompt_results.csv",
    index=False
)

print("Saved!")
```

```python
from transformers import pipeline

generator = pipeline(
    "text-generation",
    model="gpt2"
)

print("Chatbot Started")
print("Type exit to quit")

while True:

    user_input = input("\nYou: ")

    if user_input.lower() == "exit":
        break

    response = generator(
        user_input,
        max_new_tokens=100,
        do_sample=True,
        temperature=0.7
    )

    print("\nBot:")
    print(response[0]["generated_text"])
```
